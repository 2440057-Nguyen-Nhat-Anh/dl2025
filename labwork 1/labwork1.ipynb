```python
def minimize_quadratic(x, lr):
    history = []

    while x > 0.1:
        loss = x * x
        history.append(loss)
        print(f"Current loss: {x}, {loss:.4f}")
        x -= lr * 2 * x

    return history
```

```python
print("Leanrning rate = 0.1")
loss_values1 = minimize_quadratic(10, 0.1)
```

```python
print("Leanrning rate = 0.01")
loss_values2 = minimize_quadratic(10, 0.01)
```

```python
import matplotlib.pyplot as plt
plt.title("Leanrning rate = 0.1")
plt.plot(loss_values1)
plt.show()
plt.title("Leanrning rate = 0.01")
plt.plot(loss_values2)
plt.show()
```
